```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Open the image using OpenCV
image_file = "/Users/ayaanchoudhury/Desktop/Python Coding/CV-Starters/test_image.jpg"
image = cv2.imread(image_file)
```

```python
# Function to display image using matplotlib
def display(im_path):
    dpi = 80
    im_data = plt.imread(im_path)

    height, width  = im_data.shape[:2]
    
    # What size does the figure need to be in inches to fit the image?
    figsize = width / float(dpi), height / float(dpi)

    # Create a figure of the right size with one axes that takes up the full figure
    fig = plt.figure(figsize=figsize)
    ax = fig.add_axes([0, 0, 1, 1])

    # Hide spines, ticks, etc.
    ax.axis('off')

    # Display the image.
    ax.imshow(im_data, cmap='gray')

    plt.show()
```

```python
display(image_file)
```

```python
# resize the image
image_resized = cv2.resize(image, (300,300))
cv2.imwrite("/Users/ayaanchoudhury/Desktop/Python Coding/CV-Starters/test_image_resized.jpg", image_resized )
display("/Users/ayaanchoudhury/Desktop/Python Coding/CV-Starters/test_image_resized.jpg")
```

```python
# convert into grayscale
img_gray = cv2.cvtColor(image_resized, cv2.COLOR_BGR2GRAY)
cv2.imwrite("/Users/ayaanchoudhury/Desktop/Python Coding/CV-Starters/test_image_resized_grayscale.jpg", img_gray)
display("/Users/ayaanchoudhury/Desktop/Python Coding/CV-Starters/test_image_resized_grayscale.jpg")
```

```python
# blur the image
image_blur = cv2.GaussianBlur(img_gray, ((3,3)), 0)
cv2.imwrite("/Users/ayaanchoudhury/Desktop/Python Coding/CV-Starters/test_image_resized_grayscale_blur.jpg", image_blur)
display("/Users/ayaanchoudhury/Desktop/Python Coding/CV-Starters/test_image_resized_grayscale_blur.jpg")
```

```python
# canny edge detection
edges_canny = cv2.Canny(image_blur, 50,100)
cv2.imwrite("/Users/ayaanchoudhury/Desktop/Python Coding/CV-Starters/test_image_canny.jpg", edges_canny)
display("/Users/ayaanchoudhury/Desktop/Python Coding/CV-Starters/test_image_canny.jpg")
```

```python
# laplacian edge detection
edges_laplacian = cv2.Laplacian(image_blur, cv2.CV_64F)
cv2.imwrite("/Users/ayaanchoudhury/Desktop/Python Coding/CV-Starters/test_image_laplacian.jpg", edges_laplacian)
display("/Users/ayaanchoudhury/Desktop/Python Coding/CV-Starters/test_image_laplacian.jpg")
```
